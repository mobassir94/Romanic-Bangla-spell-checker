```

def is_banglish(str):
  identifier = LanguageIdentifier.from_modelstring(model, norm_probs=True)
  setofwords = set(words.words())
  en = 0
  notEn = 0
  isEn = 0
  isBn = 0
  lang = identifier.classify(str)

  if (lang[0] == 'en'):
    isEn+=1
  elif (lang[0] == 'bn'):
    isBn += 1


  if (detect(str) == 'en'):
    isEn+=1
  elif (detect(str) == 'bn'):
    isBn += 1

  strlen = len(str)
  for word in str.split():
    ck  = word in setofwords
    if (ck):
      en+=1
    else:
      notEn+=1

  en = en/strlen
  notEn = notEn/strlen

  if (en>=notEn):
    #print("En.")
    isEn+=1


  tc = nltk.classify.textcat.TextCat() 
  guess_lang = tc.guess_language(str)

  guess_country = pycountry.languages.get(alpha_3=guess_lang).name

  if (guess_country == 'English'):
    isEn+=1

  elif (guess_country == 'Bengali'):
    isBn += 1

  #print("is english? == ",isEn)
  #print("is bangla? == ",isBn)

  EnBn = isEn + isBn
  if (isEn >= 2 or isBn >= 2 or EnBn >= 2):
    return "no"
    #print("Either English or Bangla or mixure of them!")
  else:
    return str
```

```
%%time
df = pd.read_csv(comments_df[0]).text
mydf = []
count = 0
for i in range(len(df)):
  try:
    data = is_banglish(df[i])
    #print("answer ", data)
    #print(i)
    if (data != 'no'):
      mydf.append(data)
      count+=1
      print(count,data)
    if (count > 200):
       break
  except:
    pass
```
